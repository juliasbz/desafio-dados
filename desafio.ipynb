#### Importação da biblioteca Pandas

```python
import pandas as pd
```

#### Importação e atribuição dos datasets 

```python
censo = pd.read_csv("datasets/censo_estado.csv")
rouanet = pd.read_csv("datasets/rouanet.csv")
```

#### Unificação dos datasets através das colunas estado_ibge e codigo

```python
data = pd.merge(rouanet, censo, left_on='estado_ibge', right_on='codigo')
data
```

#### Troca do nome do estado para a sigla correspondente

```python
data["estado"] = data["estado"].replace({"Acre": "AC", "Alagoas": "AL", "Amapá": "AP", "Amazonas": "AM", "Bahia": "BA", "Ceará": "CE", "Espírito Santo": "ES", "Goiás": "GO", "Maranhão": "MA", "Mato Grosso": "MT", "Mato Grosso do Sul": "MS", "Minas Gerais": "MG", "Pará": "PA", "Paraíba": "PB", "Paraná": "PR", "Pernambuco": "PB", "Piauí": "PI", "Rio de Janeiro": "RJ", "Rio Grande do Norte": "RN", "Rio Grande do Sul": "RS", "Rondônia": "RO", "Roraima": "RR", "Santa Catarina": "SC", "São Paulo": "SP", "Sergipe": "SE", "Tocantins": "TO", "Distrito Federal": "DF"})
data
```

#### Remoção das linhas com valor_em_reais = 0 e quantidade = 0 

```python
data == 0
data = data[~(data == 0).any(axis=1)]
data
```

#### Criação da natural key através da soma das colunas estado_ibge e valor_em_reais

```python
data["natural_key"] = data["estado_ibge"] + data["valor_em_reais"]
data
```

#### Remoção de linhas duplicadas de acordo com a natural key

```python
data = data.drop_duplicates(subset='natural_key', keep='first')
data
```

#### Criação do arquivo final dados_tratados.csv

```python
data.to_csv("dados_tratados.csv")
```
